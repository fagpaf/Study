```python
import pandas as pd

df = pd.read_csv("consumo_energia.csv")
df.head()
import matplotlib.pyplot as plt
%matplotlib inline

# Consumo por país ao longo dos anos
for pais in df['pais'].unique():
    dados = df[df['pais'] == pais]
    plt.plot(dados['ano'], dados['consumo_GWh'], label=pais)

plt.title("Consumo de Energia Global (GWh)")
plt.xlabel("Ano")
plt.ylabel("Consumo (GWh)")
plt.legend()
plt.grid(True)
plt.show()
```
